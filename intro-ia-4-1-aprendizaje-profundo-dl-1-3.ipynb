# Aprendizaje profundo (deep learning)

El **aprendizaje profundo** es la feliz conjunción de **tres hechos**:
* El descubrimiento de **modelos** que pueden aprender mucho si hay suficientes datos (redes neuronales)
* Una gran cantidad de **datos** disponibles (vivimos en la era de los datos)
* Potencia de **cálculo** suficiente a precios asequibles (particularmente, en la nube)





En aprendizaje máquina se suele decir que **"más datos nunca molestan"**, si son representativos.
* Pero sólo del deep learning puede **aprovechar** grandes cantidades de datos, 
* Siempre que tengamos suficiente **potencia de cálculo** para entrenar el modelo.




Por cierto, siguiendo con el simil deportivo, **Rafa Nadal** sería un ejemplo de **aprendizaje profundo**: Lleva años entrenando duro, y ha sido capaz de aprovechar ese entrenamiento (Porque tiene **hardware y software de sobra**)



# Redes neuronales



* ¡La función que computa cada neurona es casi **un simple modelo lineal**! (Excepto por la función de activación)
* En esta red obtenemos el resultado según la función que ejecuta la neurona de salida, que es **una función** sencilla de otras **cuatro funciones sencillas** de otras cuatro **funciones sencillas** de las **tres variables de entrada**.
* La función de **activación** es una simple **rectificación**: si el resultado del modelo lineal es negativo, lo pone a cero. Sólo con esto las redes neuronales se adaptan perfectamente a **problemas no lineales**.
* ¿Cómo se entrena la red? De manera similar al de un modelo lineal. Por ejemplo, en un problema supervisado:
  1. Las neuronas de la **primera capa** aplican un **modelo lineal** (rectificado) a las **variables de entrada** del conjunto de **ejemplos** de entrenamiento.
  1. Las de la **siguiente capa** hacen lo mismo con los resultados de las neuronas de la capa anterior, hasta que la **última capa** calcula la **predicción** de cada ejemplo.
  1. Ahora, **hacia 'atras'** (backpropagation) se calcula la **derivada parcial** de la función de pérdida respecto a **cada parámetro** de cada neurona, modificándolos ligeramente de manera que se reduce el error medio del modelo.
  1. Y **se repite** todo el proceso hasta alcanzar (eventualmente) un **error mínimo**.

Pregunta: ¿Qué tamaño tendría una red neuronal que resolviera el problema de la predicción del vino?

## El tamaño de una red neuronal
* Se mide en las **capas** que tiene (profundidad) y el número de **neuronas** de cada capa.
* O simplemente en el **número de parámetros** que puede entrenar.

Cuanto más grande es el modelo, más cuesta entrenarlo. Hoy en día hay modelos muy grandes: **DALL-E 2** (publicado en 2022) tiene **3.5 miles de millones de parámetros** y entrenarlo costó **15 millones de dólares**.
