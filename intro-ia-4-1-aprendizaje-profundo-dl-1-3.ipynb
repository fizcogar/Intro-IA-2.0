# Redes neuronales



* La funcin que computa cada neurona es casi **un simple modelo lineal**! (Excepto por la funcin de activacin)
* En esta red obtenemos el resultado segn la funcin que ejecuta la neurona de salida, que es **una funcin** sencilla de otras **cuatro funciones sencillas** de otras cuatro **funciones sencillas** de las **tres variables de entrada**.
* La funcin de **activacin** es una simple **rectificacin**: si el resultado del modelo lineal es negativo, lo pone a cero. Slo con esto las redes neuronales se adaptan perfectamente a **problemas no lineales**.
* Cmo se entrena la red? De manera similar al de un modelo lineal. Por ejemplo, en un problema supervisado:
  1. Las neuronas de la **primera capa** aplican un **modelo lineal** (rectificado) a las **variables de entrada** del conjunto de **ejemplos** de entrenamiento.
  1. Las de la **siguiente capa** hacen lo mismo con los resultados de las neuronas de la capa anterior, hasta que la **ltima capa** calcula la **prediccin** de cada ejemplo.
  1. Ahora, **hacia 'atras'** (backpropagation) se calcula la **derivada parcial** de la funcin de prdida respecto a **cada parmetro** de cada neurona, modificndolos ligeramente de manera que se reduce el error medio del modelo.
  1. Y **se repite** todo el proceso hasta alcanzar (eventualmente) un **error mnimo**.

Pregunta: Qu tamao tendra una red neuronal que resolviera el problema de la prediccin del vino?

## El tamao de una red neuronal
* Se mide en las **capas** que tiene (profundidad) y el nmero de **neuronas** de cada capa.
* O simplemente en el **nmero de parmetros** que puede entrenar.

Cuanto ms grande es el modelo, ms cuesta entrenarlo. Hoy en da hay modelos muy grandes: **DALL-E 2** (publicado en 2022) tiene **3.5 miles de millones de parmetros** y entrenarlo cost **15 millones de dlares**.
