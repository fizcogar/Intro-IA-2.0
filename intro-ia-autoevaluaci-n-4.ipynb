**2 - Una neurona de una red neuronal implementa una función de activación $f$ aplicada a un modelo lineal, es decir: $\hat y = f(\alpha_0 + \alpha_1 \cdot  x_1 + \alpha_2 \cdot x_2 + \cdot\cdot\cdot + \alpha_n \cdot  x_n)$ ¿Qué significado tienen  los símbolos $\alpha_0, \alpha_1, \cdot\cdot\cdot, \alpha_n$ y el símbolo $\hat y$  en esta ecuación?**
1. $\alpha_0, \alpha_1, \cdot\cdot\cdot, \alpha_n$ son las variables de entrada.
1. $\hat y$ es la única entrada de la neurona.
1. $\alpha_0, \alpha_1, \cdot\cdot\cdot, \alpha_n$ son los hiperparámetros del modelo.
1. $\hat y$ es la salida de la neurona.
1. $\alpha_0, \alpha_1, \cdot\cdot\cdot, \alpha_n$ son los parámetros entrenables de la neurona.

Son ciertas la 4 y la 5. Los símbolos $\alpha_0, \alpha_1, \cdot\cdot\cdot, \alpha_n$ son los parámetros entrenables de la neurona, que se multiplican (excepto $\alpha_0$) a los valores de las variables de entrada $x_1, \cdot\cdot\cdot, x_n$.   El símbolo $\hat y$ se usa para expresar la predicción o estimación de un modelo, es decir, la salida. La y sin el 'gorro' $(y)$ se usa en cambio para expresar la magnitud correspondiente en el conjunto de datos de entrenamiento, la 'verdad verdadera', por eso un error se mide como la diferencia entre la predicción y el valor verdadero $\hat y- y$
