# La Inteligencia Artificial puede usarse para el bien y para el MAL

## Reconocimiento facial



Como cualquier tecnología, la Inteligencia Artificial puede usarse bien o  mal. El reconocimiento facial para el control de acceso de un edificio no tiene nada que ver con el reconocimiento facial de un gobierno autoritario para limitar las libertades de la población.

>[How China uses facial recognition to control human behavior](https://www.cnet.com/news/politics/in-china-facial-recognition-public-shaming-and-control-go-hand-in-hand/)
>
> "China utiliza el reconocimiento facial para perfilar a las personas de étnia Uigur, clasificarlas 
> en función de su origen étnico y señalarlas para su seguimiento, maltrato y detención."



## Algoritmos de recomendación



Los algoritmos de recomendación están diseñados con el único objeto de mantener al usuario pegado a la pantalla y aumentar así los beneficios de las compañías.

> [Enrique Dans: La idiotización de la sociedad occidental](https://www.enriquedans.com/2022/12/la-idiotizacion-de-la-sociedad-occidental.html)
>
> "¿Qué tiene TikTok que otras redes no tienen? Simplemente, un algoritmo que utiliza machine learning 
> para conseguir enganchar a sus usuarios con contenidos que los mantengan más y más tiempo delante de la pantalla."



Esto lleva a que las **fake news**, o las **teorías de la conspiración**, o las **manifestaciones de odio**, que son muy llamativas, estén entre las primeras recomendaciones.

## Sistemas para la salud



Los sistemas de IA son **más económicos** que sus homólogos basados en personas, aunque **menos precisos**. En algunos casos las empresas 'sacan de la ecuación' a los humanos para reducir costes de manera poco ética. 

Por ejemplo, en sistemas de salud privados, dejando que las IAs se encarguen de diagnosticar **con poca o ninguna supervisión humana** a los pacientes con menos recursos, y reservando la participación de médicos humanos para los pacientes más pudientes, con seguros más caros.

Pasa lo mismo con

* Empresas de recursos humanos

# ¿Qué podemos hacer?

En primer lugar, las personas que participamos en la contratación, diseño, construcción o validación (control de calidad) de sistemas de IA tenemos la **responsabilidad** de que dichos sistemas no hereden el sesgo humano. **No deja de ser un problema técnico.**



## Como administración pública, debemos:

* **Comprender y medir el impacto** del uso de la IA mediante el desarrollo y el intercambio de herramientas y enfoques
* **Ser transparente** sobre cómo y cuándo usamos la IA, comenzando con una clara **necesidad del usuario y un beneficio público**
* Proporcionar **explicaciones** significativas sobre la **toma de decisiones basadas en IA**, al tiempo que ofrece oportunidades para **revisar los resultados y cuestionar estas decisiones**
* Ser tan **abiertos** como podamos al **compartir** el código fuente y los datos de entrenamiento (protegiendo la información personal)
* Proporcionar suficiente **capacitación para los empleados públicos** que desarrollan y usan soluciones de IA y tengan las habilidades de diseño, función e implementación responsables necesarias para mejorar los servicios públicos basados en IA

[(Principios rectores del uso responsable de la inteligencia artificial (IA) (Canadá))](https://www.canada.ca/en/government/system/digital-government/digital-government-innovations/responsible-use-ai.html)

## Y, por supuesto, respetar la legislación en el uso de la IA

El parlamento europeo ha publicado una [Ley de Inteligencia Artificial](https://artificialintelligenceact.eu/)
